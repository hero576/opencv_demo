```python
from ltp import LTP
ltp = LTP()
#分句
sents = ltp.sent_split(data[1:3])
sents
```

```python
#分词
segment, hidden = ltp.seg(sents)
print(segment,hidden)
```

```python
#词性标注
pos = ltp.pos(hidden)
print(pos)
```

```python
#命名实体识别
ner = ltp.ner(hidden)
# tag, start, end = ner[0][0]
print(ner)
```

```python
#语义角色标注
srl = ltp.srl(hidden)
print(srl)
```

```python
#依存句法分析
dep = ltp.dep(hidden)
print(dep)
```

```python
#语义依存分析(树)
sdp = ltp.sdp(hidden)
print(sdp)
```
